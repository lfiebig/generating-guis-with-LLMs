```python
import pandas as pd

APP_DETAILS = "../data/app_details.csv"

UI_DETAILS = "../data/ui_details.csv"

S2W = "../data/Screen2Words.csv"
```

```python
app_details = pd.read_csv(APP_DETAILS)
app_details.drop(columns=["Icon URL"],inplace=True)

ui_details = pd.read_csv(UI_DETAILS)

s2w = pd.read_csv(S2W)
s2w.rename(columns={"screenId": "UI Number"}, inplace=True)
```

```python
details = pd.merge(app_details, ui_details, on='App Package Name')
details
```

```python
metadata = pd.merge(details, s2w[["UI Number","split"]].drop_duplicates(), on="UI Number")
metadata
```

```python
metadata.to_csv("../data/metadata.csv")
```
