```python
import pandas as pd

S2W_SAMPLE = "../data/s2w_sample.csv"

SOURCE_DIR = 'D:/unique_uis.tar/unique_uis/unique_uis/unique_uis/combined'
DESTINATION_DIR = "../data/s2w_sample1/"
```

```python
s2w_sample = pd.read_csv(S2W_SAMPLE)
s2w_sample.drop(columns=["Unnamed: 0"], inplace=True)

s2w_sample_screens = s2w_sample["screenId"]
s2w_sample_screens
```

```python
import shutil
import os

def copy_screens(screens, destination_dir = DESTINATION_DIR, source_dir = SOURCE_DIR):

    os.makedirs(destination_dir, exist_ok=True)

    for screenId in screens:
        filename = f"{screenId}.jpg"
        file_path = os.path.join(source_dir, filename)
        shutil.copy(file_path, destination_dir)
        print(f"{file_path} copied successfully")

    print("DONE.")
```

```python
copy_screens(s2w_sample_screens)
```
